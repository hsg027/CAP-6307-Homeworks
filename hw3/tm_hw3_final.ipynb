```python
#installation of pyrouge installs a function Rouge155 in the \site-packages\pyrouge directory
from pyrouge import Rouge155 #import Rouge155.py file

import os

from time import process_time # to calculate the processing time

x=Rouge155() #assigns a variable x to the function
```

```python
#assign paths of Human summaries data to the "model_dir" method defined in Rouge155 file. 
x.model_dir='C:/Users/harsh/Desktop/ucf academics/Sem-3/6307- Text Mining/hw3/HW3 (2)\HW3/Human_Summaries/eval'

#assign paths of System summaries data to the "system_dir" method defined in Rouge155 file. 
x.system_dir='C:/Users/harsh/Desktop/ucf academics/Sem-3/6307- Text Mining/hw3/HW3 (2)/HW3/System_Summaries/DPP' 

#assign the filename patter of given Human summaries data to the Rouge155 function
x.model_filename_pattern='D#ID#.M.100.T.[A-Z]'

#assign the filename patter of given System summaries data to the Rouge155 function
x.system_filename_pattern='d(\d+)t.DPP'
```

```python
#time at the start of calculation
start_time = process_time()

#store the converted and evaluated data in a local variable and print that data
dpp_output=x.convert_and_evaluate()

print(dpp_output)

# Time at the end of calculations
end_time = process_time()

print("Total time taken for calculation is:", end_time-start_time)

#Convert the ROUGE output into python dictionary for further processing.

output_dict=x.output_to_dict(dpp_output)
```
